```python
#create an instance of a AlexNet class
alexnet = models.AlexNet()
```

```python
#create an instance of a ResNet network
resnet = models.resnet101(pretrained=True)
```

```python
resnet
```

```python
my_resnet = resnet.cuda()
```

```python
#transforming the data before feeding the neural network
from torchvision import transforms
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
from PIL import Image
img = Image.open('/home/dawidkubicki/Desktop/cat.jpg')
```

```python
img.show()
```

```python
img_t = preprocess(img)
```

```python
img_t.shape
```

```python
batch_t = torch.unsqueeze(img_t, 0)
```

```python
batch_t.shape
```

```python
my_batch_t = img_t.view(1, 3, 224, 224)
```

```python
my_batch_t.shape
```

```python
my_batch_t = my_batch_t.cuda()
```

```python
#the process of running a trained model on a new data is called inference

resnet.eval() #in order to do inference
```

```python
my_out = my_resnet(my_batch_t)
```

```python
out = resnet(batch_t)
out.shape
```
